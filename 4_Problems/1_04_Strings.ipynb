## 🟩 Convert to Uppercase (1.4.1)

```python
# Code Needed
job_title = 'data scientist'
job_title_upper = job_title.upper()
print(job_title_upper)
```

## 🟩 Replace Substring (1.4.2)

```python
# Code Needed
job_title = 'Data Analyst'
job_title_replace = job_title.replace('Analyst', 'Engineer')
print(job_title_replace)

print("Data Analyst".replace('a','o'))
```

## 🟨 Concatenate Strings (1.4.3)

```python
# Code Needed
seniority_level = 'Senior'
job_title = 'Data Analyst'
full_job_title = seniority_level + ' ' + job_title
print(full_job_title)
```
