```python
# !pip install -U datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Job Posting Location Bar Plot (1.31.1)

```python
job_via_counts = df['job_via'].value_counts().head()
job_via_counts.plot(kind='bar')
plt.show()
```

---

## 🟩 Top Job Locations (1.31.2)

```python
top_locations = df['job_location'].value_counts().head()
top_locations.plot(kind='barh')
plt.xlabel('Count of Jobs')
plt.ylabel('Job Location')
plt.title('Number of Job Postings by Job Location')
plt.show()
```

---

## 🟨 Job Postings Trend Line Plot (1.31.3)

```python
df['job_posted_month'] = df['job_posted_date'].dt.to_period('M')
month_counts = df['job_posted_month'].value_counts().sort_index()
month_counts.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.title('Trend of Job Postings Over Time')
plt.show()
```
