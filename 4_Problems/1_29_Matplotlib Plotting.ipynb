```python
# !pip install -U datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Job Country Bar Chart (1.29.1)

```python
from itertools import count
import matplotlib.pyplot as plt
countries_count = df['job_country'].value_counts().head()
plt.bar(countries_count.index, countries_count)
plt.show()
```

---

## 🟨 Average Salary Horizontal Bar Chart (1.29.2)

```python
jobs_avg = df.groupby('job_title_short')['salary_year_avg'].mean().sort_values(ascending=True)
plt.barh(jobs_avg.index, jobs_avg)
plt.show()
```

---

## 🟥 Average Salary Trend Line Plot (1.29.3)

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
months_avg = df.groupby('job_posted_month')['salary_year_avg'].mean()
plt.plot(months_avg.index, months_avg)
plt.show()
```
