```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Count Job Postings by Title and Country (2.4.1)

```python
df.pivot_table(index = ['job_title_short','job_country'], aggfunc='size').head(10)
```

---

## 🟨 Salary Stats by Job Title (2.4.2)

```python
df.pivot_table(index = 'company_name', values = 'salary_year_avg', aggfunc = {'salary_year_avg': ['max', 'min', 'mean']}).head()
```

---

## 🟥 Sum Salary by Title and Country (2.4.3)

```python
df_data_scientist = df[df['job_title_short'] == 'Data Scientist']
pivot = df_data_scientist.pivot_table(index = ['company_name', 'job_country'], values = 'salary_year_avg', aggfunc = 'median')
pivot[pivot['salary_year_avg'] > 200_000].head(10)
```
