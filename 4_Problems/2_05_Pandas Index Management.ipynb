```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Set Index and Display Rows (2.5.1)

```python
df.set_index('job_posted_date', inplace=True)
df.head()
```

---

## 🟩 Reset Index and Drop Existing (2.5.2)

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

---

## 🟩 Sort Index (2.5.3)

```python
pivot = df.pivot_table(index = 'company_name', values = 'salary_year_avg', aggfunc='median')
pivot.sort_index(ascending=False, inplace=True)
pivot.head()
```

---

## 🟨 Pivot Table Reset & Set Index (2.5.4)

```python
pivot = df.pivot_table(index='job_title_short', values = 'salary_year_avg', aggfunc='median')
pivot.reset_index(inplace=True)
pivot.set_index('salary_year_avg', inplace=True)
pivot.head()
```
