```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## 🟩 Health Insurance Pie Chart (2.14.1)

```python
df_DA = df[df['job_title_short'] == 'Data Analyst']
health_insurance = df_DA['job_health_insurance'].value_counts()
health_insurance.plot(kind='pie', labels = ['No', 'Yes'], startangle=90, autopct='%1.1f%%', title='Proportion of Data Analyst Jobs Mentioning Health Insurance')
plt.show()
```

---

## 🟨 Job Schedule Distribution (2.14.2)

```python
df_ft_pt = df[df['job_schedule_type'].isin(['Full-time', 'Part-time'])]
schedule_counts = df_ft_pt['job_schedule_type'].value_counts()
plt.pie(schedule_counts, labels=['Full Time', 'Part Time'], startangle=90, autopct='%1.1f%%')
plt.title('Job Schedule Full-time vs. Part-time')


plt.show()
```

---

## 🟥 Degree Mentioned Pie Chart (2.14.3)

```python
df_US = df[df['job_country'] == 'United States']
no_deg_mention = df_US['job_no_degree_mention'].value_counts()
plt.pie(no_deg_mention, labels=['Degree Mentioned', 'No Degree Mentioned'], explode=(0, 0.1), startangle=90, autopct='%1.1f%%')
plt.title('Distribution of Jobs Mentioning Degrees for US Job Postings')
plt.show()
```
