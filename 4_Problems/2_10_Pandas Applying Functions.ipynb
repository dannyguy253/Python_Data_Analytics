```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Convert Date to String (2.10.1)

```python
df['job_posted_date_str'] = df['job_posted_date'].apply(lambda date: date.strftime('%Y-%m-%d'))
df[['job_posted_date_str', 'job_posted_date']]
```

---

## 🟨 Days Since Posted (2.10.2)

```python
from datetime import datetime

df['days_since_posted'] = df['job_posted_date'].apply(lambda date: (datetime.now() - date).days)
df[['days_since_posted', 'job_posted_date']]
```

---

## 🟥 Salary Category (2.10.3)

```python
df_filtered = df.dropna(subset='salary_year_avg').copy()
df_filtered['salary_category'] = df['salary_year_avg'].apply(lambda sal: 'Low' if sal < 60_000 else 'Medium' if sal <= 100_000 else 'High')
df_filtered[['salary_year_avg', 'salary_category']]
```
