```python
# !pip install -U datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

---

## 🟩 Sort by Salary (1.25.1)

```python
df.sort_values('salary_year_avg', ascending=False,inplace=True)
df.head(10)
```

---

## 🟩 Create Remote Column (1.25.2)

```python
df['is_remote'] = df['job_work_from_home']
```

---

## 🟩 Extract Year from Date (1.25.3)

```python
df['job_posted_year'] = pd.to_datetime(df['job_posted_date']).dt.year
df.head()
```

---

## 🟨 Drop Missing Salaries (1.25.4)

```python
df.dropna(subset = ['salary_year_avg'], inplace=True)
df.head()
```
