## 🟩 Convert Salaries to EUR (1.14.1)

```python
# Code Needed
salaries_usd = [100000, 120000, 80000, 90000]
salaries_eur = [salary * 0.85 for salary in salaries_usd]
print(salaries_eur)
```

## 🟨 Create List of Data Jobs (1.14.2)

```python
# Code Needed
job_experience = {'Data Scientist': 3, 'Data Analyst': 1, 'Machine Learning Specialist': 4, 'Data Engineer': 2}
senior_job_titles = [role for role, years in job_experience.items() if years > 2]
print(senior_job_titles)
```

## 🟥 Extract High Salary Jobs (1.14.3)

```python
# Code Needed
job_salary_dicts = [
    {'job_title': 'Data Scientist', 'salary': 100000},
    {'job_title': 'Data Analyst', 'salary': 120000},
    {'job_title': 'Data Engineer', 'salary': 80000},
    {'job_title': 'Machine Learning Engineer', 'salary': 90000}
]

high_salary_jobs = [job_salaries['job_title'] for job_salaries in job_salary_dicts if job_salaries['salary'] > 90000]
print(high_salary_jobs)
```
