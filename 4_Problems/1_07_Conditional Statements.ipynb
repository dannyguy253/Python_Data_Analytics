## 🟥 Determine Job Title (1.7.3)

```python
# Code Needed
job_experience = 4
sql_experience = 2
if (job_experience < 2):
    if (sql_experience == 0):
        print("Associate Analyst")
    else:
        print("Junior Analyst")
elif (2 <= job_experience <= 5):
    if (sql_experience == 0):
        print("Associate Analyst")
    elif (sql_experience == 1):
        print("Intermediate Analyst")
    else:
        print("Senior Analyst")
else:
    if (0 <= sql_experience <= 2):
        print("Data Analyst")
    else:
        print("Lead Analyst")
```
