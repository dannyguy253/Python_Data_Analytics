```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Drop Missing Values (2.3.1)

```python
df_copy = df.copy()
df_cleaned = df_copy.dropna(subset=['salary_year_avg'])
df_cleaned.head()
```

---

## 🟩 Random Sample of DataFrame (2.3.2)

```python
df_copy = df.copy()
df_copy = df_copy.sample(frac=0.1, random_state=42)
df_copy.head()
```

---

## 🟨 Fill Missing Salaries (2.3.3)

```python
df_copy = df.copy()
median_salary = df_copy['salary_year_avg'].median()
df_copy['salary_year_avg'] = df_copy['salary_year_avg'].fillna(median_salary)
df_copy.head()
```
