```python
# !pip install -U datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Bar Chart with Labels (1.30.1)

```python
job_counts = df['job_country'].value_counts().head(5)
plt.bar(job_counts.index, job_counts)
plt.xlabel('Country')
plt.ylabel('Number of Postings')
plt.title('Number of Jobs per Country')
plt.xticks(rotation = 45, ha='right')
plt.show()
```

---

## 🟨 Horizontal Bar Chart with Labels (1.30.2)

```python
job_title_salaries = df.groupby('job_title_short')['salary_year_avg'].mean().sort_values(ascending=True)
plt.barh(job_title_salaries.index, job_title_salaries)
plt.xlabel('Average Salary')
plt.title('Average Salary by Job Title')
plt.show()
```

---

## 🟥 Line Chart with Labels (1.30.3)

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
months_salaries = df.groupby('job_posted_month')['salary_year_avg'].mean()
plt.plot(months_salaries.index, months_salaries, color='red', label = 'Average Salary')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.title('Average Salary by Job Month')
plt.legend()
plt.show()
```
